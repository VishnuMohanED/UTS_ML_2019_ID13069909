###TITLE : GENERATIVE ADVERSARIAL NETS

Generative Adversarial Nets (GANs) is a new framework designed to generate modeling using deep learning methods i.e. Convoluted neural networks. This framework is an unsupervised learning method in machine learning in which it involves discovering and learning the patterns in the given set of data automatically and can be used to generate new samples that could have been ultimately drawn from the original dataset. This framework follows an adversarial process that eventually trains two models simultaneously. A generative model G that we train to capture data distributions, generate new samples and a discriminative model which evaluates that mostly the sample comes from training data rather than generative model. In case of any arbitrary functions G and D there exists a unique solution, where G recovers the training data distribution and D equals ½.  The entire system can be trained with back propagation when there lies the existence of back propagation.
